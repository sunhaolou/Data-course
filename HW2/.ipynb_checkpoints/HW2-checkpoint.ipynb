```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
# from mpl_toolkits import mplot3d
```

```python
def rhsfunc(x, y, epsilon):
    f1 = y[1]
    f2 = (np.square(x) - epsilon)*y[0]
    return np.array([f1, f2])

# Define some constants
xp = [-4, 4]
tol = 1e-6

# Define our initial conditions
A = 1

epsilon_start = 0 # This is our initial beta value, we will change it.

eigenfun = np.empty([81, 5])
eigenval = np.empty([1, 5])
phi = np.empty([81,5])

for modes in range(5): # Try to find 5 modes
    epsilon = epsilon_start
    depsilon = 0.05 # This is the amount we will increase epsilon by each time we don't have an eigenvalue
                 # until we get an eigenvalue

    for j in range(1000):
        y0 = np.array([A, np.sqrt((-4)**2 - epsilon)*A])
        x_evals = np.linspace(-4, 4, 20*4+1)
        sol = scipy.integrate.solve_ivp(lambda x,y: rhsfunc(x, y, epsilon), xp, y0, method = 'RK45', t_eval = x_evals)
        y_sol = sol.y[0, :]
        yprime = sol.y[1, :]
        error = yprime[-1] + np.sqrt(4**2 - epsilon) * y_sol[-1]

        if np.abs(error)<tol:
            area = np.sqrt(scipy.integrate.trapz(sol.y[0, :]**2, x = x_evals))
            norm = scipy.integrate.trapz((sol.y[0, :]/area)**2, x = x_evals)
            print(norm)
            eigenfun[:, modes] = np.abs(sol.y[0, :]/area).T
            eigenval[:, modes] = epsilon
            phi[:, modes] = np.abs(sol.y[0, :]/area).T
            break

        if (-1)**(modes)*error > 0: 
            epsilon = epsilon + depsilon
        else:
            epsilon = epsilon - depsilon/2
            depsilon = depsilon/2

    epsilon_start = epsilon + 0.1

    plt.plot(sol.t, sol.y[0, :]/area, linewidth=2)
    plt.plot(sol.t, 0*sol.t, 'k')
```

```python
def rhsfunc2(x, y, epsilon, gamma):
    f1 = y[1]
    f2 = (gamma*np.square(y[0]) + np.square(x) - epsilon)*y[0]
    return np.array([f1, f2])

# Define some constants
xp = [-3, 3]
tol = 1e-5
gamma = 0.05
x_evals = np.linspace(-3, 3, 20*3+1)
# Define our initial conditions
A_init = 1e-3 # This is the shooting-method parameter that we will change
epsilon_start = 0 # This is our initial beta value, we will change it.

eigenfun = np.empty([61, 2])
eigenval = np.empty([1, 2])

for modes in range(2):
    epsilon = epsilon_start 
    depsilon = 0.05
    A = A_init
    for j in range(1000):
        y0 = np.array([A, np.sqrt((-3)**2 - epsilon)*A])
        sol = scipy.integrate.solve_ivp(lambda x,y: rhsfunc2(x, y, epsilon, gamma), xp, y0, method = 'RK45', t_eval = x_evals)
        y_sol = sol.y[0, :]
        yprime = sol.y[1, :]
        error = yprime[-1] + np.sqrt(3**2 - epsilon) * y_sol[-1]
        area = np.sqrt(scipy.integrate.trapz(sol.y[0, :]**2, x = x_evals))
        norm = scipy.integrate.trapz(np.square(sol.y[0, :]), x = x_evals)
        if np.abs(norm-1) < tol and np.abs(error)<tol:
            eigenfun[:, modes] = np.abs(sol.y[0, :]/area).T
            eigenval[:, modes] = epsilon
            break
        
        else:
            A = A/np.sqrt(norm)
        
        y0 = np.array([A, np.sqrt((-3)**2 - epsilon)*A])
        sol = scipy.integrate.solve_ivp(lambda x,y: rhsfunc2(x, y, epsilon, gamma), xp, y0, method = 'RK45', t_eval = x_evals)
        y_sol = sol.y[0, :]
        yprime = sol.y[1, :]
        error = yprime[-1] + np.sqrt(3**2 - epsilon) * y_sol[-1]
        area = np.sqrt(scipy.integrate.trapz(sol.y[0, :]**2, x = x_evals))
        norm = scipy.integrate.trapz(np.square(sol.y[0, :]), x = x_evals)
        if np.abs(norm-1) < tol and np.abs(error)<tol:
            eigenfun[:, modes] = np.abs(sol.y[0, :]/area).T
            eigenval[:, modes] = epsilon
            break
        else:
            if (-1)**(modes)*error > 0: 
                epsilon = epsilon + depsilon
            else:
                epsilon = epsilon - depsilon/2
                depsilon = depsilon/2

    epsilon_start = epsilon + 0.1
```
