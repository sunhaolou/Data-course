```python
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.integrate
```

```python
f = lambda t, y: -3*y*np.sin(t) # dydt = f(t,y)
f_true = lambda t: math.pi*math.exp(3*(np.cos(t)-1))/math.sqrt(2)
```

```python
def FE(tn, yn, delta_t):
    yn_plus_1 = yn + delta_t*f(tn, yn)
    return yn_plus_1
```

```python
delta_t = np.array([2**(-2), 2**(-3), 2**(-4), 2**(-5), 2**(-6), 2**(-7), 2**(-8)])
```

```python
def y5_FE(delta_t):
    result = []
    t = 0
    y = math.pi/math.sqrt(2)
    result.append(y)
    while t < 5:
        yn_plus_1 = FE(t, y, delta_t)
        result.append(yn_plus_1)
        y = yn_plus_1
        t = t + delta_t
    return result
```

```python
y_N_FE = np.zeros(7)
```

```python
for i in range(7):
    y_N_FE[i] = y5_FE(delta_t[i])[-1]
```

```python
y_true = f_true(5)
error_FE = np.zeros(7)
for i in range(7):
    error_FE[i] = abs(y_true - y_N_FE[i])
```

```python
plt.loglog(delta_t, error_FE)
```

```python
slope_FE, intercept_FE = np.polyfit(np.log(delta_t), np.log(error_FE), 1)
```

```python
A1 = np.reshape(y5_FE(delta_t[-1]), (1281, 1))
A2 = np.reshape(error_FE, (1, 7))
A3 = slope_FE
```

```python
np.set_printoptions(precision = 16)
```

```python
print(A1[-1])
print(A2)
print(A3)
print(y_true)
```

```python
def Heun_method(tn, yn, delta_t):
    yn_plus_1 = yn + delta_t/2*(f(tn, yn) + f(tn+delta_t, yn+delta_t*f(tn,yn)))
    return yn_plus_1
```

```python
def y5_Heun(delta_t):
    result = []
    t = 0
    y = math.pi/math.sqrt(2)
    result.append(y)
    while t < 5:
        yn_plus_1 = Heun_method(t, y, delta_t)
        result.append(yn_plus_1)
        y = yn_plus_1
        t = t + delta_t
    return result
```

```python
y_N_Heun = np.zeros(7)
```

```python
for i in range(7):
    y_N_Heun[i] = y5_Heun(delta_t[i])[-1]
```

```python
y_true = f_true(5)
error_Heun = np.zeros(7)
for i in range(7):
    error_Heun[i] = abs(y_true - y_N_Heun[i])
```

```python
plt.loglog(delta_t, error_Heun)
```

```python
slope_Heun, intercept_Heun = np.polyfit(np.log(delta_t), np.log(error_Heun), 1)
```

```python
A4 = np.reshape(y5_Heun(delta_t[-1]), (1281, 1))
A5 = np.reshape(error_Heun, (1, 7))
A6 = slope_Heun
```

```python
print(A4.shape)
print(A5)
print(A6)
```

```python
def Adams(tn, yn_1, yn, delta_t):
    y_p = yn + delta_t/2*(3*f(tn, yn) - f(tn-delta_t, yn_1))
    yn_plus_1 = yn + delta_t/2*(f(tn+delta_t, y_p) + f(tn, yn))
    return yn_plus_1
```

```python
def y5_Adams(delta_t):
    result = []
    t = delta_t
    yn_1 = math.pi/math.sqrt(2)
    y = yn_1 + delta_t*f(0+delta_t/2, yn_1+delta_t/2*f(0, yn_1))
    result.append(yn_1)
    result.append(y)
    while t < 5:
        yn_plus_1 = Adams(t, yn_1, y, delta_t)
        result.append(yn_plus_1)
        yn_1 = y
        y = yn_plus_1
        t = t + delta_t
    return result
```

```python
y_N_Adams = np.zeros(7)
```

```python
for i in range(7):
    y_N_Adams[i] = y5_Adams(delta_t[i])[-1]
```

```python
y_true = f_true(5)
error_Adams = np.zeros(7)
for i in range(7):
    error_Adams[i] = abs(y_true - y_N_Adams[i])
```

```python
plt.loglog(delta_t, error_Adams)
```

```python
slope_Adams, intercept_Adams = np.polyfit(np.log(delta_t), np.log(error_Adams), 1)
```

```python
A7 = np.reshape(y5_Adams(delta_t[-1]), (1281, 1))
A8 = np.reshape(error_Adams, (1, 7))
A9 = slope_Adams
```

```python
print(A7.shape)
print(A8)
print(A9)
```

```python
plt.loglog(delta_t, error_FE, 'o', label='FE')
plt.loglog(delta_t, (delta_t**slope_FE)*np.exp(intercept_FE), label='slope_FE')
plt.loglog(delta_t, error_Heun, 'o', label='Heun')
plt.loglog(delta_t, (delta_t**slope_Heun)*np.exp(intercept_Heun), label='slope_Heun')
plt.loglog(delta_t, error_Adams, 'o', label='Adams')
plt.loglog(delta_t, (delta_t**slope_Adams)*np.exp(intercept_Adams), label='slope_Adams')
plt.xlabel('log($\Delta$t)')
plt.ylabel('log(error)')
plt.title('Error vs. delta_t on log-log plot for all three methods')
plt.legend()
plt.savefig('error_loglog_plot.png')
plt.show()
```

```python
from scipy.integrate import solve_ivp
```

```python
def vdp(t, x):
    y = x[0]
    v = x[1]
    return [v, epsilon*(1-y**2)*v - y]
```

```python
epsilon = 0.1
t = np.arange(0, 32+0.5, 0.5)
sol_first = solve_ivp(fun=vdp, t_span=[t[0], t[-1]], y0=[math.sqrt(3), 1], method='RK45', t_eval=t)
```

```python
epsilon = 1
t = np.arange(0, 32+0.5, 0.5)
sol_second = solve_ivp(fun=vdp, t_span=[t[0], t[-1]], y0=[math.sqrt(3), 1], method='RK45', t_eval=t)
```
